## 합성곱 (convolution)
- 인공 신경망
  - 처음에 가중치 $w_1$ ~ $w_{10}$과 절편 $b$를 랜덤하게 초기화한 다음 에포크를 반복하면서 경사 하강법 알고리즘을 사용하여 손실이 낮아지도록 최적의 가중치와 절편을 찾아감
  - 출력의 개수는 입력 개수와 상관없이 뉴런의 개수와 동일
- 합성곱
  - 입력 데이터 전체에 가중치를 적용하는 것이 아니라 일부에 가중치를 곱함
  - 뉴런이 한 칸씩 이동하여 가중치와 특성을 곱해 새로운 출력을 만듦
  - **첫 번째 합성곱에 사용된 가중치 $w_1$ ~ $w_{10}$과 절편 $b$가 두 번째 합성곱에도 동일하게 사용됨**
  - 합성곱 층의 뉴런에 있는 가중치 개수는 하이퍼파라미터
- 뉴런을 **필터**또는 **커널**이라고 부름
  - 필터: 뉴런의 개수를 이야기할 때
  - 커널: 입력에 곱해지는 가중치를 의미할 때
- 합성곱의 장점: 1차원이 아니라 2차원 입력에도 적용 가능
  - 마치 도장을 찍듯이 왼쪽 위에서 오른쪽 맨 아래까지 이동하면서 출력을 만듦
  - 입력과 가중치의 행과 열에 맞추어 곱셈하고 모두 더함
  - **특성 맵**(feature map): 합성곱 계산을 통해 얻은 출력
    - ⚠️ 특성 맵은 활성화 함수를 적용한 후의 값
  - ⚠️ 합성곱 층에 있는 각 필터의 가중치(커널)는 모두 다름 (사진에서 $w1$이 모두 같은 값이 아님!!)
  - 👉 2차원 구조를 그대로 사용하기 때문에 합성곱 신경망이 이미지 처리 분야에서 뛰어난 성능을 발휘

## 케라스 합성곱 층
```python
from tensorflow import keras
keras.layers.Conv2D(10, kernel_size=(3, 3), activation='relu')
```
- 첫 번째 매개변수: 필터의 개수
- `kernel_size`: 필터에 사용할 커널의 크기를 정함 (보통 `(3,3)`이나 `(5,5)` 권장)
- 1개 이상의 합성곱 층을 쓴 인공 신경망을 합성곱 신경망이라고 부름 (꼭 합성곱 층만 사용한 신경망을 합성곱 신경망이라고 부르는 것은 아님!)
### 패딩과 스트라이드
#### 패딩 (padding)
- 커널 크기는 그대로 두고 입력과 동일한 크기의 출력(특성맵)을 만들기 위해서는 마치 더 큰 입력에 합성곱하는 척 해야함
![img](https://velog.velcdn.com/images/jujaemin/post/490671c4-b9a3-4cfe-99a3-f50eaf3d3a78/image.JPG)
- 패딩: 입력 배열의 주위를 가상의 원소로 채우는 것
  - 패딩은 0으로 채움
  - 👉 계산에 영향을 미치지는 않음
- 세임 페딩 (same padding): 입력과 특성 맵의 크기를 동일하게 만들기 위해 입력 주위에 0으로 패딩하는 것 (합성곱 신경망에서 많이 사용)
![img](https://velog.velcdn.com/images/jeonghyunje/post/0a3d82f2-7b3e-4c1e-8aed-43d6dfe6a61f/image.png)
  - ⭐**세임 패딩의 경우 입력과 만들어진 특성 맵의 가로세로 크기가 동일함**
- 밸리드 패딩 (valid padding): 패딩 없이 순수한 입력 배열에서만 합성곱을 하여 특성맵을 만드는 경우 (특성 맵의 크기가 줄어들 수 밖에 없음)
  - 모서리에 있는 중요한 정보가 특성 맵으로 잘 전달되지 않을 가능성이 높음, 반면 가운데 있는 정보는 두드러지게 표현됨
- 적절한 패딩은 이미지 주변에 있는 정보를 잃어버리지 않도록 도와줌

![img](https://velog.velcdn.com/images%2Fboing-86%2Fpost%2Fb2f1947e-d4b8-4c7d-8643-b84d92db11bd%2F%E1%84%89%E1%85%B3%E1%84%8F%E1%85%B3%E1%84%85%E1%85%B5%E1%86%AB%E1%84%89%E1%85%A3%E1%86%BA%202022-02-24%20%E1%84%8B%E1%85%A9%E1%84%92%E1%85%AE%207.20.04.png)
```python
# padding 파라미터 기본값은 valid
keras.layers.Conv2D(10, kernel_size=(3, 3), activation='relu', padding='same')
```
#### 스트라이드 (stride)
- 이동의 크기
```python
# strides 파라미터 기본값은 1
keras.layers.Conv2D(10, kernel_size=(3, 3), activation='relu', padding='same', strides=1)
```
- 1보다 큰 스트라이드를 사용하는 경우는 드묾 👉 `strides` 매개변수는 잘 사용하지 않음
### 풀링 (pooling)
- 합성곱 층에서 만든 특성 맵의 가로세로 크기를 줄이는 역할
![img](https://velog.velcdn.com/images/jeonghyunje/post/c8eeba62-f437-4eae-9244-d02a40078559/image.png)
- ⭐**특성 맵의 개수는 줄어들지 않음!** (`(2, 2, 3)` ➡️ `(1, 1, 3)`)
- 합성곱처럼 입력 위를 지나가면서 도장을 찍지만, 가중치가 없음
- ⭐**가로세로 방향으로만 진행**
- 최대 풀링 (max pooling): 도장을 찍은 영역에서 가장 큰 값을 고름
  - 많은 경우 평균 풀링보다 최대 풀링을 더 많이 사용
![img](https://velog.velcdn.com/images/jeonghyunje/post/83b02f2d-c989-42b3-9053-ad650f16dc04/image.png)
- 평균 풀링 (average pooling): 도장을 찍은 영역에서 평균값을 계산
  - 특성 맵에 있는 중요한 정보를 (평균하여) 희석시킬 수 있음
- 풀링에서는 겹치지 않고 이동함 👉 풀링 크기와 스트라이드가 동일(풀링의 크기가 `(2, 2)`라면 스트라이드의 크기도 `2`)
- 패딩 없음
```python
keras.layers.MaxPooling2D(2, stides=2, padding="valid")
```
  - 첫 번째 매개변수: 풀링의 크기 지정
    - 대부분의 풀링의 크기는 2 (가로세로 크기를 절반으로 줄임)
  - `strides`와 `padding` 매개변수를 제공하지만, `strides` 값은 자동으로 풀링의 크기이고 `padding`은 자동으로 `valid`이므로 이 매개변수를 바꾸는 경우는 거의 없음
  - `AveragePooling2D`: 평균 풀링

## 합성곱 신경망의 전체 구조
![img](https://velog.velcdn.com/images/jeonghyunje/post/1b66978d-15ce-420c-86e2-ada46a372e94/image.png)
- 1️⃣ 합성곱 층
  - 필터의 가중치는 각기 서로 다름
  - 따로 언급하지 않는다면 합성곱의 스트라이드는 항상 1
  - 세임 패딩으로 만들어진 특성 맵의 크기는 입력과 동일한 `(4, 4)`
  - 합성곱 층에서도 활성화 함수를 적용함 (렐루 함수를 많이 사용)
- 2️⃣ 풀링 층
  - 합성곱 층에서 만든 특성 맵의 가로세로 크기를 줄임
  - 보통 `(2, 2)` 풀링을 사용하여 절반으로 줄임
  - 특성 맵의 개수는 변하지 않음
  - 풀링을 사용하는 이유: 합성곱에서 스트라이드를 크게 하여 특성 맵을 줄이는 것보다 풀링 층에서 크기를 줄이는 것이 경험적으로 더 나은 성능
  - 합성공 신경망은 합성곱 층에서 특성 맵을 생성하고 풀링에서 크기를 줄이는 구조가 쌍을 이룸
- 3️⃣ 밀집 층
  - 3차원 배열을 1차원으로 펼침
  - 3개의 클래스를 분류하는 다중 분류 문제 👉 소프특스 활성화 함수 사용
