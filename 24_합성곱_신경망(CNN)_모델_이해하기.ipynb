## í•©ì„±ê³±ì¸µ - Conv2D
- íŒ¨ë”©(padding): ì´ë¯¸ì§€ì˜ ê°€ì¥ìë¦¬ì— ë¹ˆ ê³µê°„(0)ì„ ì¶”ê°€í•´ ì´ë¯¸ì§€ ê°€ì¥ìë¦¬ ë¶€ë¶„ì˜ í”½ì…€ì´ ì¶©ë¶„íˆ ì²˜ë¦¬ë  ìˆ˜ ìˆë„ë¡ ê¸°íšŒë¥¼ ë¶€ì—¬  
  - ì„¸ì„ íŒ¨ë”©(same padding): ì…ë ¥ ì£¼ìœ„ì— ì ì ˆí•œ ê°œìˆ˜ì˜ íŒ¨ë”©ì„ ì¶”ê°€ ğŸ‘‰ **ì…ë ¥ê³¼ íŠ¹ì„± ë§µì˜ í¬ê¸°ê°€ ë™ì¼**
  - ë°¸ë¦¬ë“œ íŒ¨ë”©(valid padding): íŒ¨ë”© ì—†ì´ ìˆœìˆ˜í•œ ì…ë ¥ ë°°ì—´ì— ëŒ€í•´ì„œë§Œ í•©ì„±ê³±ì„ ìˆ˜í–‰ ğŸ‘‰ ì…ë ¥ì˜ í¬ê¸°ë³´ë‹¤ ì‘ì€ íŠ¹ì„±ë§µì˜ í¬ê¸°
- ìŠ¤íŠ¸ë¼ì´ë“œ(stride): í•„í„°ê°€ ì´ë¯¸ì§€ ìœ„ë¥¼ ì´ë™í•˜ëŠ” ì†ë„
- `keras.layers.Conv2D()`
  - 2ì°¨ì› í•©ì„±ê³± ìˆ˜í–‰
  - í•„ìˆ˜ ë§¤ê°œë³€ìˆ˜: í•„í„° ê°œìˆ˜, ì»¤ë„ í¬ê¸°

```python
import keras
from keras import layers

conv1 = layers.Conv2D(filters=10, kernel_size=(3, 3))
# kernel_size: 3ì°¨ì› í•©ì„±ê³±ì´ë¼ë©´ ì„¸ ê°œì˜ ì •ìˆ˜ íŠœí”Œ
```

- `filters`: í•©ì„±ê³±ì¸µ ì¶œë ¥ì˜ ë§ˆì§€ë§‰ ì°¨ì› ê²°ì •
- `kernel_size`: í•„í„°ì˜ ì²˜ìŒ ë‘ ì°¨ì›ì¸ ë„ˆë¹„ì™€ ë†’ì´

```python
import numpy as np

x = np.random.normal(size=(10, 28, 28, 1))
conv_out = conv1(x)
print(conv_out.shape)
```

- ì²« ë²ˆì§¸: ë°°ì¹˜(batch)ì˜ ì°¨ì›
- ë‘ ë²ˆì§¸, ì„¸ ë²ˆì§¸: ë†’ì´ì™€ ë„ˆë¹„
- ë„¤ ë²ˆì§¸: ì±„ë„ í˜¹ì€ íŠ¹ì„± ë§µì˜ ê¹Šì´

- í•„í„°ì˜ ê¹Šì´ == ì…ë ¥ì˜ ê¹Šì´
- í•„í„°ì˜ ê°œìˆ˜ == íŠ¹ì„± ë§µì˜ ê¹Šì´
  - í•©ì„±ê³±ì¸µì´ ê±°ë“­ë  ìˆ˜ë¡ ë§ì€ í•„í„°ë¥¼ ì‚¬ìš©í•˜ëŠ” ê²½í–¥ ğŸ‘‰ íŠ¹ì„± ë§µì˜ ê¹Šì´ëŠ” ì ì  ê¹Šì–´ì§
- íŒ¨ë”©ì„ ì¶”ê°€í•˜ì§€ ì•Šìœ¼ë©´ í•©ì„±ê³± ì—°ì‚°ì´ ê³µê°„ ë°©í–¥ì˜ ì°¨ì›ì„ ì¤„ì„
- ğŸ‘‰ **íŠ¹ì„± ë§µì˜ ê¹Šì´ê°€ ê¹Šì–´ì§€ë©´ì„œ ëª¨ë¸ì´ ì ì  ë” ë³µì¡í•œ ì •ë³´ì™€ ì„¸ë¶€ì ì¸ íŒ¨í„´ì„ ì´í•´í•˜ê²Œ ë§Œë“¤ê³ , ì´ë¯¸ì§€ì˜ ê³µê°„ ë°©í–¥ì˜ ì°¨ì›ì´ ì¤„ì–´ë“¤ê²Œ ë§Œë“¤ì–´ ì ì  ì¤‘ìš”í•œ íŠ¹ì„±ë§Œ ë‚¨ê¸°ëŠ” ê³¼ì •!**

```python
# ìŠ¤íŠ¸ë¼ì´ë“œ 2ë°°
conv2 = layers.Conv2D(filters=10, kernel_size=3, strides=(2, 2))
print(conv2(x).shape)
```

```python
# strides=2, same padding
conv3 = layers.Conv2D(filters=10, kernel_size=3, strides=2,
                      padding="same")
print(conv3(x).shape)
```

```python
# strides=1, same padding
conv4 = layers.Conv2D(filters=10, kernel_size=3, padding="same")
print(conv4(x).shape)
```

## í’€ë§ì¸µê³¼ ë°€ì§‘ì¸µ - AveragePooling2D, Dense
### Pooling
- ì…ë ¥ëœ í…ì„œì˜ ë†’ì´ì™€ ë„ˆë¹„ë¥¼ ì¤„ì´ë©´ì„œ í”½ì…€ ê°’ì„ ì••ì¶•í•˜ëŠ” ê¸°ëŠ¥
- ì´ë¯¸ì§€ì˜ í¬ê¸°ë¥¼ ì¤„ì´ë©´ì„œ í•©ì„±ê³±ì¸µì—ì„œ ì¶”ì¶œëœ ì¤‘ìš” ì •ë³´ë¥¼ ì••ì¶•í•˜ëŠ” ê³¼ì •
- ì²« ë²ˆì§¸ ë°°ì¹˜ ì°¨ì›ê³¼ ë§ˆì§€ë§‰ ì°¨ì›ì„ ì œì™¸í•œ ë‚˜ë¨¸ì§€ ì°¨ì›ì˜ í¬ê¸°ë¥¼ ì¤„ì„
- í’€ë§ ìœˆë„(pooling window)ê°€ ì…ë ¥ í…ì„œ ìœ„ë¥¼ ì§€ë‚˜ê°€ë©´ì„œ í‰ê· ê°’ì„ ê³„ì‚°í•˜ê±°ë‚˜ (í‰ê·  í’€ë§ average pooling) ìµœëŒ“ê°’ì„ ì„ íƒ (ìµœëŒ€ í’€ë§ max pooling)
- í’€ë§ í¬ê¸°, ìŠ¤íŠ¸ë¼ì´ë“œ, íŒ¨ë”© ì§€ì • ê°€ëŠ¥
  - ìŠ¤íŠ¸ë¼ì´ë“œì˜ í¬ê¸°ê°€ í•­ìƒ í’€ë§ì˜ í¬ê¸°ì™€ ë™ì¼
  - íŒ¨ë”©ì€ ê±°ì˜ í•­ìƒ "valid"

```python
pool1 = layers.AveragePooling2D(pool_size=2)
pool2 = layers.AveragePooling2D(pool_size=3)

print(pool1(x).shape)
print(pool2(x).shape)
```

### Dense
- ì…ë ¥ê³¼ ê°€ì¤‘ì¹˜ì˜ ì ê³±(dot product) ìˆ˜í–‰ ($y = x \times W + b$)

```python
# ìœ ë‹›ì´ 3ê°œ, ì…ë ¥ íŠ¹ì„±ì´ 2ê°œì¸ ë°€ì§‘ì¸µ
dense1 = layers.Dense(3)

x2 = np.array([[5, 7]])
print(dense1(x2).shape)
```

![img](https://velog.velcdn.com/images/j_zero/post/32ba9cf0-635b-4841-b7ff-daf38db5cf92/image.png)

```python
print(dense1.get_weights())

dense1.set_weights([
    np.array([[1, 2, 3], [4, 5, 6]]),
    np.array([0, 0, 0])
])

print(dense1(x2))
```

```python
weight = dense1.get_weights()[0]
print(np.dot(x2, weight))
```
