### AlexNet ëª¨ë¸ ë§Œë“¤ê¸°
- LeNet-5 ëª¨ë¸ë³´ë‹¤ ë§ì€ ì¸µì„ ì‚¬ìš©
- LeNet-5 ëª¨ë¸ì—ì„œ ì‚¬ìš©í•œ í™œì„±í™” í•¨ìˆ˜ì¸ ì‹œê·¸ëª¨ì´ë“œ í•¨ìˆ˜ ëŒ€ì‹  ë ë£¨ í•¨ìˆ˜ ì‚¬ìš©
  - **ê·¸ë ˆì´ë””ì–¸íŠ¸ ì†Œì‹¤ ë¬¸ì œ**(gradient vanishing problem): ì‹œê·¸ëª¨ì´ë“œ í•¨ìˆ˜ëŠ” ì…ë ¥ ê°’ì´ ì»¤ì§€ê³  ì‘ì•„ì§ì— ë”°ë¼ í•¨ìˆ˜ì˜ ê¸°ìš¸ê¸°ê°€ ê¸‰ê²©í•˜ê²Œ ì¤„ì–´ë“¦ ğŸ‘‰ ê¸°ìš¸ê¸°ê°€ ì‘ë‹¤ë©´ íŒŒë¼ë¯¸í„°ë¥¼ ì—…ë°ì´íŠ¸í•˜ëŠ” ë° ê±°ì˜ ì˜í–¥ì„ ë¯¸ì¹˜ì§€ ëª»í•¨
- í‰ê·  í’€ë§ ëŒ€ì‹  ìµœëŒ€ í’€ë§ ì‚¬ìš©
  - ìŠ¤íŠ¸ë¼ì´ë“œ í¬ê¸°ê°€ í’€ë§ í¬ê¸°ë³´ë‹¤ ì‘ì•„ ì¤‘ì²©ëœ í’€ë§ì„ ìˆ˜í–‰ ğŸ‘‰ ìµœëŒ€ í’€ë§ì„ ì‚¬ìš©í•˜ë©´ ì¤‘ìš” ì •ë³´ë¥¼ í¬ì„ì‹œí‚¤ì§€ ì•ŠìŒ
- ë°€ì§‘ì¸µì˜ ê³¼ëŒ€ì í•© ë°©ì§€ë¥¼ ìœ„í•´ ë“œë¡­ì•„ì›ƒ ì‚¬ìš©

```python
import keras
from keras import layers

alexnet = keras.Sequential()
alexnet.add(layers.Input(shape=(227, 227, 3))) # ì»¬ëŸ¬ ì´ë¯¸ì§€

# 2ê°œì˜ í•©ì„±ê³±ì¸µê³¼ í’€ë§ì¸µ
alexnet.add(layers.Conv2D(filters=96, kernel_size=11, strides=4,
                          activation='relu'))
alexnet.add(layers.MaxPooling2D(pool_size=3, strides=2))
alexnet.add(layers.Conv2D(filters=256, kernel_size=5, padding='same',
                          activation='relu'))
alexnet.add(layers.MaxPooling2D(pool_size=3, strides=2))

# 3ê°œì˜ í•©ì„±ê³±ì¸µ
alexnet.add(layers.Conv2D(filters=384, kernel_size=3, padding='same',
                          activation='relu'))
alexnet.add(layers.Conv2D(filters=384, kernel_size=3, padding='same',
                          activation='relu'))
alexnet.add(layers.Conv2D(filters=256, kernel_size=3, padding='same',
                          activation='relu'))

# 1ê°œì˜ í’€ë§ì¸µ
alexnet.add(layers.MaxPooling2D(pool_size=3, strides=2))

# 2ê°œì˜ ë°€ì§‘ì¸µê³¼ ë“œë¡­ì•„ì›ƒì¸µ
alexnet.add(layers.Flatten())
alexnet.add(layers.Dense(units=4096, activation='relu'))
alexnet.add(layers.Dropout(rate=0.5))
alexnet.add(layers.Dense(units=4096, activation='relu'))
alexnet.add(layers.Dropout(rate=0.5))

# ì¶œë ¥ì¸µ
alexnet.add(layers.Dense(1000, activation="softmax"))
```

## ì‚¬ì „ í›ˆë ¨ëœ CNN ëª¨ë¸ - VGGNet
- ë°˜ë³µì  ë¸”ë¡ êµ¬ì¡°ë¼ëŠ” ê°œë…ì„ ì•ˆì°©ì‹œí‚¨ ëª¨ë¸
- ì—¬ëŸ¬ ë²ˆì˜ í•©ì„±ê³±ì¸µ ë‹¤ìŒì— í’€ë§ì¸µì„ ì ìš©í•˜ëŠ” ë°©ì‹ì„ ì±„íƒ

### VGGNet ëª¨ë¸ ë§Œë“¤ê¸°
- í•©ì„±ê³±ì¸µê³¼ í’€ë§ì¸µì„ êµëŒ€ë¡œ ë°˜ë³µí•˜ëŠ” ëŒ€ì‹ , ì—¬ëŸ¬ ë²ˆì˜ í•©ì„±ê³±ì¸µì„ ì ìš©í•œ ë‹¤ìŒ í’€ë§ì¸µì„ ì ìš©
  - ğŸ‘‰ ê¹Šì€ ë„¤íŠ¸ì›Œí¬ë¥¼ ë§Œë“¤ê¸° ìœ„í•´ í’€ë§ì„ ì ìš©í•˜ê¸° ì „ ì—¬ëŸ¬ ê°œì˜ í•©ì„±ê³±ì¸µì„ ë°°ì¹˜
- ë™ì¼í•œ ì—¬ëŸ¬ ê°œì˜ í•©ì„±ê³±ì¸µê³¼ í’€ë§ì¸µì˜ êµ¬ì¡°ë¥¼ ë°˜ë³µí•˜ëŠ” **ë¸”ë¡**ì„ ì ìš©
- í° í•„í„°ë¥¼ ì‚¬ìš©í•˜ëŠ” í•˜ë‚˜ì˜ í•©ì„±ê³±ì¸µ ëŒ€ì‹ , 3 X 3 í¬ê¸°ì˜ ì‘ì€ í•„í„°ë¥¼ ì‚¬ìš©í•˜ëŠ” ì—¬ëŸ¬ ê°œì˜ í•©ì„±ê³±ì¸µì„ ì ìš©

```python
# VGG16

vggnet = keras.Sequential()
vggnet.add(layers.Input(shape=(224, 224, 3)))

# 1, 2ë²ˆì§¸ ë¸”ë¡: í•©ì„±ê³±ì¸µ 2ê°œ, í’€ë§ì¸µ 1ê°œ
for n_filters in [64, 128]:
  for _ in range(2):
    vggnet.add(layers.Conv2D(filters=n_filters, kernel_size=3,
                             padding='same', activation='relu'))
  vggnet.add(layers.MaxPooling2D(pool_size=2))

# 3, 4ë²ˆì§¸ ë¸”ë¡: í•©ì„±ê³±ì¸µ 3ê°œ, í’€ë§ì¸µ 1ê°œ
for n_filters in [256, 512, 512]:
  for _ in range(3):
    vggnet.add(layers.Conv2D(filters=n_filters, kernel_size=3,
                             padding='same', activation='relu'))
  vggnet.add(layers.MaxPooling2D(pool_size=2))


vggnet.add(layers.Flatten())
vggnet.add(layers.Dense(4096, activation='relu'))
vggnet.add(layers.Dense(4096, activation='relu'))
vggnet.add(layers.Dense(1000, activation='softmax'))
```
